```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
otp = pd.read_csv('data\historikus_2018_02_16__2019_02_15_.csv')
otp
```

```python
otp['Utolsó ár'].plot()
```

```python
N=len(otp)
N
```

```python
otp['returns'] = np.log(otp['Utolsó ár']/otp['Utolsó ár'].shift(1))
```

```python
otp.returns.plot()
```

```python
sns.histplot(otp.returns,stat='count', kde=True)
```

```python
otp.returns.describe()
```

```python
sns.histplot(otp.returns,stat='frequency')
```

```python
sns.histplot(otp.returns,stat='density', binwidth=0.005, kde=True)
```
